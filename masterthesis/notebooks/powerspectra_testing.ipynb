```python
import os
import sys
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

# add path
parent_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
# parent_dir = os.path.dirname(os.path.realpath(__file__))
sys.path.append(parent_dir)
# parent_dir


from src.features import classPK 
```

```python
redshift = 10.0
Pk_synchronous = classPK.ClassSpectra(redshift, 'synchronous')
Pk_newtonian = classPK.ClassSpectra(redshift, 'newtonian')
```

```python
phi_sync = Pk_synchronous.phi_pk
phi_newt = Pk_newtonian.phi_pk

plt.loglog(phi_sync.k, phi_sync.pk, label='synchronous')
plt.loglog(phi_newt.k, phi_newt.pk, label='newtonian')
plt.show()
```

```python
pk_sync = Pk_synchronous.d_tot_pk
pk_newt = Pk_newtonian.d_tot_pk

```

```python
plt.loglog(pk_sync.k, pk_sync.pk, label='synchronous')
plt.loglog(pk_newt.k, pk_newt.pk, label='newtonian')
plt.legend()
plt.show()
```

```python
# inferred from gravitational potential

pk_sync_grav = Pk_synchronous.pk_frame 
pk_newt_grav = Pk_newtonian.pk_frame
```

```python
plt.loglog(pk_sync_grav.k, pk_sync_grav.pk, label='synchronous')
plt.loglog(pk_newt_grav.k, pk_newt_grav.pk, label='newtonian')
plt.legend()
plt.show()
```
