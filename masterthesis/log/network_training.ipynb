## Explanations

### Create Dataloader

The dataloaders are created from a subset of the available seeds. For instance we may use 1750 seeds for training, testing and validation and the remaining 250 seeds to evaluate model performance. This works because the seeds are all random initialisations drawn from the same distribution. For each seed we have one datacube per gravitation theory per redshift, which are not independent of each other. 

Both dataloaders are generated from a subset consisting of the 1745 first seeds of the collection. These are then divided at random into:
 * train-loader (70%)
 * test-loader (20%)
 * validation-loader (10%).

#### Whole cubes

The [whole cube dataloaders](../src/data/whole_cube_dataset.py) are loaded as follows:

```python
WC_train, WC_test, WC_val = whole_cube_dataset.make_whole_dataset(
    train_test_val_split = (0.7, 0.2, 0.1),
    batch_size=8,
    num_workers=4,
    redshifts=1.0,
    total_seeds=np.arange(0,1750,1),
    prefetch_factor=2,
)
```

```python
# Sample information - one sample is one cube
print(f"{len(WC_train.dataset)} training samples")
print(f"{len(WC_test.dataset)} testing samples")
print(f"{len(WC_val.dataset)} validation samples")
print(f"{len(WC_train.dataset)+len(WC_test.dataset)+len(WC_val.dataset)} total samples\n")

# Look at one batch
batch1 = next(iter(WC_train))
data = batch1["image"]
label = batch1["label"]
print(f"Looking at the first batch:\nThere are {len(data)} images in this batch, with labels: \n{label}")

# Second batch
batch2 = next(iter(WC_train))
data = batch2["image"]
label = batch2["label"]
print(f"Looking at the second batch:\nThere are {len(data)} images in this batch, with labels: \n{label}")
```

#### Sliced cubes

The [sliced cube dataloaders](../src/data/sliced_cube_dataset.py) are loaded as follows:

```python
SL_train, SL_test, SL_val = sliced_cube_dataset.make_sliced_dataset(
    train_test_val_split = (0.7, 0.2, 0.1),
    batch_size=256*3,
    num_workers=4,
    redshifts=1.0,
    total_seeds=np.arange(0,1750,1),
    prefetch_factor=2,
)

batchSL = next(iter(SL_train))
data = batchSL["image"]
label = batchSL["label"]
```

```python
fig, axes = plt.subplots(ncols=3, nrows=3, figsize=(10,10), sharex=True, sharey=True)
plt.rcParams['image.origin'] = 'lower'
idx = 0
for i in range(3):
    for j in range(3):
        axes[i,j].imshow(data[idx][0])
        axes[i,j].set_title(f"Label: {label[idx][0]}")
        idx += 34
# Colorbar 
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
fig.colorbar(axes[0,0].imshow(data[0][0]), cax=cbar_ax)
fig.suptitle("Sliced Cubes", fontsize=16)
```
